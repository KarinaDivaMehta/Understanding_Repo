Uniform random variable â†’ All outcomes are equally likely

Expected value of a continuous uniform random variable using integral function

$ \text{If x} \in [a,b] \text{ then} $  
$ \text{PDF f(x)} = \frac{1}{b-a} $

$$
\mathbb{E}[X] = \int_a^b x f(x)\,dx \\[0.5em]
\mathbb{E}[X] = \int_a^b x \frac{1}{b-a}\,dx \\[0.5em]
\text{Factor out the constants} \\[0.5em]
\mathbb{E}[X] = \frac{1}{b-a} \int_a^b x\,dx \\[0.5em]
\text{Power rule of integration: } \int x^n dx = \frac{x^{n+1}}{n+1} \\[0.5em]
\mathbb{E}[X] = \frac{1}{b-a} \left[ \frac{x^2}{2} \right]_a^b \\[0.5em]
\mathbb{E}[X] = \frac{1}{b-a} \left( \frac{b^2}{2} - \frac{a^2}{2} \right) \\[0.5em]
\mathbb{E}[X] = \frac{b^2 - a^2}{2(b-a)} \\[0.5em]
\mathbb{E}[X] = \frac{(b-a)(b+a)}{2(b-a)} \\[0.5em]
\mathbb{E}[X] = \frac{b+a}{2} \\[0.5em]
\text{Since } a=0,\; b=1 \\[0.5em]
\mathbb{E}[X] = \frac{1}{2}
$$

- Consider making a new variable, $\varepsilon = X - \mathbb{E}[X]$
- What's the expectation of $\varepsilon$?


$$
\text{Suppose } \\[0.5em]
 \varepsilon = X - \mathbb{E}[X]. \\[0.5em]

\text{Expectation of } \varepsilon: \\[0.5em]
\mathbb{E}[\varepsilon] = \mathbb{E}[X - \mathbb{E}[X]]  \\[0.5em]
= \mathbb{E}[X] - \mathbb{E}[X] = 0 \\[0.5em]
$$

- What's the variance of $\varepsilon$?


$$
\text{Variance of } \varepsilon: \\[0.5em]
\mathbb{V}[\varepsilon] = \mathbb{E}\big[(\varepsilon - \mathbb{E}[\varepsilon])^2\big] \\[0.5em]
= \mathbb{E}[\varepsilon^2] \\[0.5em]
= \mathbb{E}[(X - \mathbb{E}[X])^2] \\[0.5em]
= \mathbb{V}[X] \\[0.5em]
$$

- So we can write any random variable in the form $X = \mathbb{E}[X] + \varepsilon, $ where $\mathbb{E}[\varepsilon]=0$ and $\mathbb{V}[\varepsilon] = \sigma_X^2$
- If that's true, show that we can also write any random variable in the form $X = \mathbb{E}[X] + \sigma_X \varepsilon$, where $\mathbb{E}[\varepsilon]=0$ and $\mathbb{V}[\varepsilon]=1$


$$
\text{new variable:} \\
\tilde{\varepsilon} = \frac{\varepsilon}{\sigma_X}. \\[1mm]

\text{Expectation of } \tilde{\varepsilon} \\
\mathbb{E}[\tilde{\varepsilon}] = \mathbb{E}\left[\frac{\varepsilon}{\sigma_X}\right] 
= \frac{1}{\sigma_X} \mathbb{E}[\varepsilon] = 0 \\[1mm]

\text{Variance of } \tilde{\varepsilon} \\
\mathbb{V}[\tilde{\varepsilon}] = \mathbb{V}\left[\frac{\varepsilon}{\sigma_X}\right]
